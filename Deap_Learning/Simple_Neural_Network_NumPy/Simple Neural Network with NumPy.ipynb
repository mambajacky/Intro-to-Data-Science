# Simple Linear Neural Network with NumPy

Build a linear model using numpy and train the model using L-2 norm loss function and gradient descent optimization

### Import relevant libraries 

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

### Generate random input data and target

```python
observations = 1000
x = np.random.uniform(low=-10, high=10, size=(observations,1))
y = np.random.uniform(low=-10, high=10, size=(observations,1))
inputs = np.column_stack((x,y)) #join 1-D arrays to a single 2-D array
```

```python
noise = np.random.uniform(-1, 1, (observations,1))
targets = 2 * x - 3* y + 5 + noise
```

### Plot training data

```python
targets = targets.reshape(observations,)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, targets)
ax.view_init(azim=50)
targets = targets.reshape(observations,1)
plt.show()
```

### Initiate variables 

```python
init_range = 1
weights = np.random.uniform(-init_range, init_range, (2,1))
biases = np.random.uniform(-init_range, init_range, 1)

print(weights, biases)
```

### Set a learning rate 

```python
learning_rate = 0.05
```

### Train the model 

```python
weights_new = weights
biases_new = biases

for i in range(200):
    outputs = np.dot(inputs, weights_new) + biases_new
    deltas = outputs - targets
    
    # loss function, it changes nothing to scale it
    loss = np.sum(deltas ** 2) / 2 / observations
    print(loss)
    
    # gradient descent
    weights_new = weights_new - learning_rate * np.dot(inputs.T, deltas/observations)
    biases_new = biases_new - learning_rate * np.sum(deltas/observations)
```

### Print the weights and biases 

They are fairly close to the actual values

```python
print(weights_new, biases_new)
```

### Plot targets with outputs 

```python
plt.plot(targets, outputs)
plt.xlabel('target', fontsize=15)
plt.ylabel('outputs', fontsize=15)
plt.show()
```
